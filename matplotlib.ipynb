# Animations

Animations require relating the time of your problem to the frame of the animation, then specifying the frames per second
For isntance --> a sin wave moving at 3m/s:

```python
frame = 50

def function(x,t):
    return np.sin(x - 3*t)

x = np.linspace(0, 10*np.pi, 1000)

###############################################
from matplotlib import animation
from matplotlib.animation import PillowWriter
###############################################

fig, ax = plt.subplots(1, 1, figsize=(8,4))
ln1, = plt.plot([],[])
time_text = ax.text(0.65, 0.95, '', fontsize=15, transform=ax.transAxes, bbox=dict(facecolor='white', edgecolor='black'))

ax.set_xlim(0, 10*np.pi)
ax.set_ylim(-1.5, 1.5)

def animate(i):
    ln1.set_data(x, function(x, 1/frame*i))
    time_text.set_text('t={:.2f}'.format(i/frame))

ani = animation.FuncAnimation(fig, animate, frames=240, interval=frame)
ani.save('animation.gif', writer='pillow', fps=frame, dpi=100)
```

Dynamic 3D Plot

```python
plt.style.use(['default'])

fig, ax = plt.subplots(subplot_kw={'projection':'3d'})

ax.plot_surface(x,y,z, cmap='coolwarm', linewidth=0, antialiased=False)
ax.view_init(elev=10, azim=0)


def animate(i):
    ax.view_init(elev=10, azim=3*i) # every frame moves 3 degrees


ani = animation.FuncAnimation(fig, animate, frames=120, interval=50)
ani.save('animation2.gif', writer='pillow', fps=30, dpi=100)
```

Plotting All Data from a File in a Single Figure

```python
################################################
plt.style.use(['science', 'notebook', 'grid'])
################################################
############# Hyper Parameters ################# 
################################################
normalized = True                   # Normalizes all intesnities to 1
FigSize = (14, 10)                  # Changes your Figure size
x_value = 'wavelength [nm]'         # Sets the x axis label
y_value = 'Intenity [arb. units]'   # Sets the y axis label
addLable = True                     # Sets a lable to the plots and takes the file name for that
y_extend = 1.5                      # If the lable intersects with the plot, you can increase this value to prevent this from happening (only works with Norm.)
labelFont = 10                      # Changes the Fontsize of the lable
tickFontSize = 12
###############################################

dimension = len(data)

if dimension % 2 !=0:

    dimension = len(data) + 1

    fig, axes = plt.subplots(int(dimension/2), 2, figsize=FigSize)

    count = 0

    for i in range(int(dimension/2)):
        for j in range(2):
            ax = axes[i][j]
            ew, intensity = np.loadtxt(data[count], unpack=True)

            if normalized: 
                intensity = intensity/np.max(intensity)
                ax.set_ylim(top=y_extend)
                
            ax.plot(ew, intensity, label= os.path.basename(data[count]).replace('.csv', ''))
            ax.tick_params(axis='both', labelsize=tickFontSize)


            if addLable:
                ax.legend(loc='upper right', fontsize=labelFont) 

            if count >= len(data)-1:
                ax.set_xlabel(x_value)

            if count % 2 == 0:
                print('yuhu')
                ax.set_ylabel(y_value)

            count += 1
               
else:
    fig, axes = plt.subplots(int(dimension/2), 2, figsize=FigSize) 

    count = 0

    for i in range(int(dimension/2)):
        for j in range(2):
            ax = axes[i][j]
            ew, intensity = np.loadtxt(data[count], unpack=True)

            if normalized: 
                intensity = intensity/np.max(intensity)
                ax.set_ylim(top=y_extend)

            ax.plot(ew, intensity, label= os.path.basename(data[count]).replace('.csv', ''))
            ax.tick_params(axis='both', labelsize=tickFontSize)

            if addLable:
                ax.legend(loc='upper right', fontsize=labelFont) 
            
            if count >= len(data)-2:
                ax.set_xlabel(x_value)

            if count % 2 == 0:
                ax.set_ylabel(y_value)

            count += 1
```
